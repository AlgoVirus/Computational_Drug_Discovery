## Search for Target protein

### Target search for coronavirus

```python
# Target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

### Select and retrieve bioactivity data for SARS coronavirus 3C-like proteinase(seventh entry)
We will assign the seventh entry (which corresponds to the target protein, coronavirus 3C-loke proteinase) to the ***selected_target***

```python
selected_target = targets.target_chembl_id[6]
selected_target
```

here, we will retrieve only bioactivity data for  coronavirus 3C-like proteinase(CHEMBL3927) that are reported as IC50 values in nM (nanomolar) unit.


```python
activity = new_client.activity
res =  activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

# Data pre-processing of bioactivity data

#### Labeling compound as either being active, inactive or intermediate
The bioactivity data is the IC50 unit, Compounds having values of less than 1000nM will be considered to be **active** while those greater than 10,000 nM will be considered to be **inactive**. As for those values in between 1,000 and 10,000 nM will be refered to as **intermediate**.

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

#### Iterate the *molecule_chembl_id, canoniacl_smiles and standard_value* to a list

```python
selection = ['molecule_chembl_id','canonical_smiles', 'standard_value']
df3 = df2[selection]
```

```python
df3
```

```python
bioactivity_class = pd.Series(bioactivity_class, name='bioactivity_class')
df4 =  pd.concat([df3, bioactivity_class], axis=1)
df4
```

Saves dataframe to  CSV file

```python
df4.to_csv('bioactivity_data_preprocessed.csv', index=False)
```
